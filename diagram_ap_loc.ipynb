```python
from Analyzer import Analyzer
import numpy as np
```

```python
analyzer = Analyzer()
analyzer.loadData()
analyzer.dataset = analyzer.dataset.fillna(-85)
```

```python
numerical_labels = np.array([analyzer.repository.locations.keys().index(tuple(l)) for l in analyzer.labels])
numerical_labels = map(int, numerical_labels)
analyzer.dataset['location',] = numerical_labels

# count amount of different locations we have
loc_amount = len(set(numerical_labels))
print "LocAmount: ", loc_amount
locbin = {}
for location in range(0, loc_amount):
    locbin[location] = 0
    # reduce dataset.
    location_dataset = analyzer.dataset.loc[lambda dataset: dataset.location == location, :]

    for column in location_dataset:
        if column != "location":
            if max(location_dataset[column].tolist()) > -85:
                locbin[location] += 1
    print location, " : ", locbin[location]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(locbin.values())
plt.xlabel('location')
plt.ylabel('number of AP > 85db')
plt.show()
```
